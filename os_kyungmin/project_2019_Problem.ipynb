### 서로다른 3자리숫자를 맞추는 프로그램

```python
#### 필요한 패키지 import
import random
secret=["1","2","3","4","5","6","7","8","9"]

#### 위 리스트에 있는 문자(숫자로 된)들을 랜덤하게 섞은 리스트를 변수 secret에 저장한다.
random.shuffle(secret)
print(secret)
print(secret[:3])

#### 앞에서부터 3개의 문자를 연결하여 3자리 문자(예를 들면 "357")로 만들어 secret_number에 저장
secret_number = "".join(secret[:3])
print(secret_number)
```

```python
print(secret_number)

number_of_trial = 0 # 시도 횟수
progress = ''       # 인쇄할 메세지
max_num = 10        # 최대 시도 횟수
while number_of_trial < max_num:
    flag = False
    exist = 0
    correct = 0
    x = input('서로 다른 3자리 숫자:')
    
# 입력된 데이터가 3자리 인지 체크하는 블럭
    #### if 문: 만일 입력된 데이터가 3자리가 아니면
    if not len(x) is 3:
        print('3자리가 아닙니다, 다시 입력하세요.')
        print(progress)
        continue
        
# 입력된 데이터가 모두 숫자인지 체크하는 블럭
    #### if 문: 만일 입력된데이터가 모두 숫자인 경우가 아니면
    if not x.isdecimal():
        print('숫자가 아닌 입력이 있습니다. 다시 입력하세요.')
        print(progress)
        continue
        
# 입력된 데이터 3개가 모두 다른지 체크하는 블럭 
    for i in range(3):
        #### if문: 만일 이 자리의 숫자가 이 전에 나온 숫자들중 하나라도 같으면
        if x[i] in x[:i]:
            flag = True
            break
            
    if flag:
        print('같은 숫자가 있습니다. 다시 입력하세요.')
        print(progress)
        continue

# 입력한 숫자와 secret_number와 몇개가 일치 하는지 체크하는 블럭
    #### 각 입력한 숫자마다
    for elem in x:
        #### secret_number의 어느 위치에 있는 숫자와도 일치하면
        if elem in secret_number:
            #### exist 값을 1 증가
            exist += 1

# 위치와 값이 일치하는 숫자가 몇개인지 체크하는 블럭
    #### 각 입력한 숫자마다
    for i in range(len(x)):
        #### 입력숫자 위치와 동일한 위치에 있는 secret_number의 숫자와 일치하면
        if x[i] == secret_number[i]:
            #### correct 값을 1 증가
            correct += 1
            
# 모두 맞추었는지 체크하는 블럭
    if correct == 3:
        print(progress)
        print('비밀번호를 맞추었습니다:',secret_number)
        print(number_of_trial+1,'번 시도 했습니다.')
        break
        
# 진행상황 업데이트하는 블럭
    progress = progress + str(x) + '\n위치는 틀리고 비밀번호와 값만 일치한 숫자 갯수: '+\
    str(exist - correct)+'\n위치와 값 모두 일치하는 숫자 갯수: '+str(correct) + '\n'
    print(progress)
    number_of_trial += 1
    
# while 루프를 벗어난 후, 졌는지를 체크하는 블럭
if number_of_trial > max_num:
    print('졌습니다!!. \n최대 횟수를 초과 했습니다.')
    print('비밀번호:',secret_number)
        
```
